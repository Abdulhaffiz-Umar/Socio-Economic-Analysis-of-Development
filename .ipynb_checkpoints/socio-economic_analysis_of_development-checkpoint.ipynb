## Introduction

This project will examine the environmental and socio-economic characteristics of developed and developing countries using the World Bank's country classification model<sup><a href="#1">1</a></sup>. The model classifies countries according to their Gross National Income (GNI) per capita. Findings from this report will help understand how sustainable development, urban population growth, CO2 emissions, and human development index differ between developed and developing countries.

Gapminder's dataset will be used to conduct the analysis. Gapminder is an independent Swedish foundation with no political, religious or economic affiliations. Gapminder is a fact tank, not a think tank. Gapminder fights devastating misconceptions about global development<sup><a href="#2">2</a></sup>.

## Materials and Methods

### Data Sources

The following data was obtained from Gapminder's website:

- **GNI per Capita**<br>
The GNI (Gross National Income) is the sum of a country’s gross domestic product (GDP) plus net income (positive or negative) from abroad.<br>
Source: https://www.gapminder.org/data/ 


- **CO2 emissions per person**<br>
Carbon dioxide emissions from the burning of fossil fuels (Metric tonnes of C02 per person).<br>
Source: https://cdiac.ess-dive.lbl.gov/ 


- **Sustainable development index** <br>
The sustainable development is an efficiency metric designed to access the ecological efficiency of nations to deliver human development. <br>
Source: http://gapm.io/dsdi


- **Urban population growth**<br>
Urban population refers to people living in urban areas as defined by national statistical offices. It is calculated using World bank's population estimates and urban ratios from the United Nations World Urbanization Prospects.<br>
Source: https://data.worldbank.org/indicator/SP.URB.GROW


- **Life expectancy**<br>
The average number of years a newborn child would live if current mortalitiy patterns were to stay the same. .<br>
Source: http://gapm.io/ilex


- **Income per person**<br>
Gross domestic product per person adjusted for differences in purchasing power.<br>
Source: http://gapm.io/dgdppc


- **Total population**<br>
Source: http://gapm.io/dpop


- **Geographic Regions**<br>
Data on the six geographic regions of the world<br>
Source: https://www.gapminder.org/data/geo/

## Data Wrangling

### Import Libraries 

> Before we begin analysis, it is necessary to import the libraries we will use for the analysis. 


```python
# Import necessary packages 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina' 
```

### Load Data


```python
#Load GNI per capita data 
df_gni = pd.read_csv('gnipercapita_ppp_current_international.csv')
df_gni.head()
```

```python
df_gni.shape
```

```python
#load CO2 emissions data
df_co2 = pd.read_csv('co2_emissions_tonnes_per_person.csv')
df_co2.head()
```

```python
df_co2.shape
```

```python
#load Sustainable development index data
df_sdi = pd.read_csv('sdi.csv')
df_sdi.head()
```

```python
df_sdi.shape
```

```python
# load urban population growth data
df_urban = pd.read_csv('urban_population_growth_annual_percent.csv')
df_urban.head()
```

```python
df_urban.shape
```

```python
#load life expectancy data
df_life = pd.read_csv('life_expectancy_years.csv')
df_life.head()
```

```python
df_life.shape
```

```python
#load incomme data
df_inc = pd.read_csv('income_per_person_gdppercapita_ppp_inflation_adjusted.csv')
df_inc.head()
```

```python
df_inc.shape
```

### Melt Data
Change the structure of dataframes using "melt" function

```python
# melt gni data
df_gni_m = pd.melt(df_gni, id_vars=["country"], var_name="year", value_name= "gni_per_capita")

#sort values by country and year
df_gni_m.sort_values(["country", "year"] , inplace = True)
#show the head of the dataframe
df_gni_m.head()
```

```python
# melt co2 data
df_co2_m = pd.melt(df_co2, id_vars=["country"], var_name="year", value_name= "co2_emission")

#sort values by country and year
df_co2_m.sort_values(["country", "year"] , inplace = True)
#show the head of the dataframe
df_co2_m.head()
```

```python
#melt sdi data
df_sdi_m = pd.melt(df_sdi, id_vars=["country"], var_name="year", value_name= "sdi")

#sort values by country and year
df_sdi_m.sort_values(["country", "year"] , inplace = True)
#show the head of the dataframe
df_sdi_m.head()
```

```python
# melt urban population data
df_urban_m = pd.melt(df_urban, id_vars=["country"], var_name="year", value_name= "urban_population")

#sort values by country and year
df_urban_m.sort_values(["country", "year"] , inplace = True)
#show the head of the dataframe
df_urban_m.head()
```

```python
# melt life expectancy data
df_life_m = pd.melt(df_life, id_vars=["country"], var_name="year", value_name= "life_expectancy")

#sort values by country and year
df_life_m.sort_values(["country", "year"] , inplace = True)
#show the head of the dataframe
df_life_m.head()
```

```python
# melt income data
df_inc_m = pd.melt(df_inc, id_vars=["country"], var_name="year", value_name= "income")

#sort values by country and year
df_inc_m.sort_values(["country", "year"] , inplace = True)
#show the head of the dataframe
df_inc_m.head()
```

### Add region and population data

```python
df_region = pd.read_csv('geography.csv')
df_region.head()
```

```python
df_pop = pd.read_csv('population.csv')
df_pop.head()
```

```python
df_pop.dtypes
```

```python
# convert year to str, else we can't perform the merge later
df_pop['year'] = df_pop['year'].astype(str)
```

```python
df_pop.dtypes
```

### Merge the dataset into one dataframe
Use the "merge" function to combine the dataset into a single dataframe

```python
# merge the data on 'country' and 'year' column

df_join1 = pd.merge(pd.merge(df_gni_m, df_sdi_m, on = ['country', 'year'], how = 'inner'), 
            df_co2_m, how = 'inner', on = ['country', 'year']) #join the first three dataset together
df_join2 = pd.merge(pd.merge(df_join1, df_urban_m, on = ['country', 'year'], how = 'inner' ), 
            df_life_m, on = ['country', 'year'], how = 'inner') #join  next two other dataset     
df_join3 = pd.merge(df_join2, df_inc_m, on = ['country', 'year'], how = 'inner') #join the last  melted dataset
df_join3.head()
```

```python
# merge 'geograpy' and 'population' dataset to the dataframe

df_join4 = pd.merge(pd.merge(df_join3, df_region, on = 'country', how = 'inner' ), 
            df_pop, on = ['country', 'year'], how = 'inner')
df_join4.head()
```

```python
cols = list(df_join4.columns.values)
cols
```

```python
# rearrange the columns 
df_join4 = df_join4[['country','year','region','population','gni_per_capita','sdi','co2_emission','urban_population','life_expectancy','income']]
df_join4.head()
```

```python
#save the merged dataframe 
df_join4.to_csv('gen_data.csv', index = False)
```

### Convert 'k' currency values to integers
I observed that the 'gni_per_capita' and 'income' column has 'k' currency values, I will convert this into a numerals for better analysis


```python
# fix gni_per_capita column
df_gen_data['gni_per_capita'] = df_gen_data['gni_per_capita'].str.replace('k', ' * 1000') # multiplies k into 1000
df_gen_data['gni_per_capita'] = df_gen_data['gni_per_capita'].map(pd.eval).astype(int) # populates the result across the column
```

```python
# fix income column
df_gen_data['income'] = df_gen_data['income'].str.replace('k', ' * 1000') # multiplies k into 1000
df_gen_data['income'] = df_gen_data['income'].map(pd.eval).astype(int) # populates the result across the column
```

## Exploratory Data Analysis

### Research Question 1: What is the relationship between life expectancy and income?
> We shall examine the relationship between life expectancy and income by making bubble chart consisting of life expectancy and income variables for the year 2018 and 1990 respectively. The chart is adapted from Hans Rosling's Factfulness. First, let's prepare the data by adding a color column, then plot the bubble chart

```python
# load data
df_data = pd.read_csv('gen_cleaned.csv')
df_data.head()
```

#### create a colour column to represent each region

```python
# create a color pallete to match each region
df_data.region.unique()
```

```python
color_pallete = {
    "europe_central_asia":'#f49d37', #orange
    "middle_east_north_africa": '#5be56b', #green
    "sub_saharan_africa": '#4e7af0', #blue
    "america": '#ebcc21', #gold
    "east_asia_pacific": '#35d1d1', #teal
    "south_asia": '#f03838', #red
}
```

```python
df_data["color"] = df_data.region.apply(lambda x: color_pallete[x])
```

```python
df_data
```

```python
# select only 2018 data 
df_data_18 = df_data[df_data.year == 2018].copy()
```

```python
# select only 1990 data 
df_data90 = df_data[df_data.year == 1990].copy()
```

```python
df_data_18.head()
```

```python
df_data90.head()
```

#### Create plot for 2018
Next, the plot will be created with life expectancy on the 'y' axis and income per capita on the 'x' axis. The population will represent the bubble size

```python
fig, ax = plt.subplots(figsize = (18,12))
# increase the font size
plt.rcParams.update({'font.size': 12})

x = df_data_18["income"]
y = df_data_18["life_expectancy"]
color = np.array(df_data_18["color"])
size = df_data_18["population"]

#plot scatter
plt.scatter (x = x, y = y, s = size/70000, c = color, label = None, edgecolors = "black")


#set labels
ax.set_ylabel ("Life Expectancy")
ax.set_xlabel ("Income Per Person")
ax.set_title ("Relationship Between Life Expectancy and Income in 2018")

#remove the side bars to make it look better
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# reduce y axis to allow for better visualization
ax.set_ylim(50,90)

#add legend
#Source: #https://jakevdp.github.io/PythonDataScienceHandbook/04.06-customizing-legends.html3

for region in list(df_data_18.region.unique()):
    plt.scatter([], [], c=color_pallete[region], alpha=1, label=str(region), s = 100, edgecolors = "black")
    
legend1 = plt.legend(scatterpoints=1, frameon=False, labelspacing=1, loc =4, title='')

plt.show()
```

From the plot, we can observe that Africa and South Asia are still lagging behind in development. This is relative to the two continent being regarded as regions with the least development prospects.<sup><a href="#4">4</a></sup> On the other hand, America and Europe Central Asia are well advanced with high income and longer life expectancy. 

#### Create plot for 1990

```python
fig, ax = plt.subplots(figsize = (18,12))
# increase the font size
plt.rcParams.update({'font.size': 12})

x = df_data90["income"]
y = df_data90["life_expectancy"]
color = np.array(df_data90["color"])
size = df_data90["population"]

#plot scatter
plt.scatter (x = x, y = y, s = size/70000, c = color, label = None, edgecolors = "black")


#set labels
ax.set_ylabel ("Life Expectancy")
ax.set_xlabel ("Income Per Person")
ax.set_title ("Relationship Between Life Expectancy and Income in 1990")

#remove the side bars to make it look better
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# reduce y axis to allow for better visualization
ax.set_ylim(40,90)

#add legend

for region in list(df_data90.region.unique()):
    plt.scatter([], [], c=color_pallete[region], alpha=1, label=str(region), s = 100, edgecolors = "black")
    
legend1 = plt.legend(scatterpoints=1, frameon=False, labelspacing=1, loc =4)

plt.show()
```

### Research Question 2: How does developed countries and developing countries compare in terms of CO2 emissions?
what is the trend in carbon emissions in developed and developing countries from 1990 to 2018? To answer this, we will plot a line chat to check which region emits more carbon. 


#### Create a column for country profile
We need a column that will separate countries into developed and developing countries. To do this. we will use the World Bank's country classification for 2018<sup><a href="#5">5</a></sup> which classified countries into low-income, lower-middle income, upper-middle income, and high-income respectively. 


| Threshold   | GNI/Capita     |
| ----------- | -------------- |
| Low-income  | < 995          |
| Lower-middle income | 996 - 3,895    |
| Upper-middle income | 3,896 - 12,055 |
| High income | > 12,055               |


Further, low- and middle-income economies are referred to as developing economies, and the Upper Middle Income and the High Income are referred to as Developed Countries<sup><a href="#6">6</a></sup>, therefore, we shall categorize the countries according to their gni_per_capital in relation to World Bank's metrics. 

```python
df_dataCP = df_data.copy()
```

```python
df_dataCP.head()
```

```python
# define a function that groups countries based on their gni_per_capita
def categorise(df_dataCP):
    if df_dataCP['gni_per_capita'] <= 3895: # for developing countries with low gni
        return 'developing'
    elif df_dataCP['gni_per_capita'] > 3895: # for developed countries with high gni
        return 'developed'
```

```python
df_dataCP['country_profile'] = df_dataCP.apply(lambda df_dataCP: categorise(df_dataCP), axis =1)
df_dataCP.head()
```

```python
df_developing = df_dataCP[df_dataCP["country_profile"] == "developing"] # dataframe for developing countries 
df_developed =df_dataCP[df_dataCP["country_profile"] == "developed"]  # dataframe for developedountries 
```

```python
df_developed.to_csv('whats_wrong.csv')
```

#### Plot line chart

```python
# plot the line chart

# increase the chat size
fig, ax = plt.subplots(figsize = (12,6)) 

#add grid style
sns.set_style('whitegrid')

# increase the font size
plt.rcParams.update({'font.size': 12})

sns.lineplot(data = df_developed, x = "year", y = "co2_emission", ci = None, label = 'developed countries')
sns.lineplot(data = df_developing, x = "year", y = "co2_emission", ci = None, label = 'developing countries')


ax.set_title ("Trend in co2 emissions \n 1990 - 2018" , pad = 20)
ax.legend(loc = 'lower right' , bbox_to_anchor=(1.25, 0), frameon=False);
```

The graph shows a slight downtrend in co2 emissions for both regions, athough it is apparent that developed countries emit significantly more carbon than developing countries over the time period. 

### Research Question 3: Comparison between developing and developed countries in terms of co2 emissions and urban population

In this section, we shall observe the relationship between urban population and co2 emissions in developing countries and developed countries via a regression plot.

```python
df_developed.describe()
```

> I noticed that urban population contains non-numeric characters and won't evaluate, to fix this, we will use the "to_numeric method  with a coerce argument.<sup><a href="#7">7</a></sup>

```python
df_developed['urban_population'] = pd.to_numeric(df_developed['urban_population'], errors='coerce')
df_developing['urban_population'] = pd.to_numeric(df_developing['urban_population'], errors='coerce')
```

```python
#check if it worked
df_developed['urban_population'].dtypes
```

```python
# it worked
df_developed.describe()
```

```python
#create a correlation factor matrix
df_developed[['gni_per_capita', 'sdi', 'co2_emission', 'urban_population']].corr()
```

```python
#create a correlation factor matrix
df_developing[['gni_per_capita', 'sdi', 'co2_emission', 'urban_population']].corr()
```

#### Plot scatterplot

```python
#Plot co2 emissions x urban population
fig, ax = plt.subplots(figsize = (10, 10)) 

#plot the chart
sns.regplot(data =  df_developed, x = 'urban_population' , y = 'co2_emission', fit_reg = True, scatter_kws = {'alpha':0.6}, label = 'developed countries')
sns.regplot(data =  df_developing, x = 'urban_population', y = 'co2_emission', fit_reg = True, scatter_kws = {'alpha':0.6}, label = 'developing countries');

ax.set_title('Correlation between CO2 emissions and Urban Population' , pad = 30)
ax.set_xlabel('Urban population growth')
ax.set_ylabel('Tonnes of CO2 emissions per capita')
ax.grid(False) 
ax.legend(loc = 'lower right' , bbox_to_anchor=(1.30, 0), frameon=False)


plt.show()
```

From the plot, we observe that there is a positive correlation between co2 emissions and urban population in developed countries while there is a negative correlation between the variables for developing countries. 

 #### Correlation between economy and Co2 emissions

```python
#Plot co2 emissions x gni_per_capita
fig, ax = plt.subplots(figsize = (12, 12)) 

#plot the chart
sns.regplot(data =  df_developed, x = 'gni_per_capita' , y = 'co2_emission', fit_reg = True, scatter_kws = {'alpha':0.5}, label = 'developed countries', ax = ax)
sns.regplot(data =  df_developing, x = 'gni_per_capita', y = 'co2_emission', fit_reg = True, scatter_kws = {'alpha':0.5}, label = 'developing countries', ax = ax)

ax.set_title('Correlation between Economic Development and Co2 Emissions' , pad = 30)
ax.set_xlabel('GNI/Capita adjusted for inflation')
ax.set_ylabel('Tonnes of CO2 emissions per capita')
ax.grid(False) 
ax.legend(loc = 'lower right' , bbox_to_anchor=(1.30, 0), frameon=False)


plt.show()
```

There exist a strong positive correlation between co2 emissions and gni per capita in developed countries, the dataset seems similar for developing countries, however, this is an anomaly that will be presented as a limitation in this study.

### Research Question 4: Relationship between sustainable development and urban population
Here, we seek to know whether there is a relationship between urban population and sustainable development, does urban urban population stimulate sustainable development? We will investigate this by finding the correlation between both factors. 

```python
#Plot sdi x gni_per_capita
fig, ax = plt.subplots(figsize = (10, 10)) 

#plot the chart
sns.regplot(data =  df_developed, x = 'sdi' , y = 'urban_population', fit_reg = True, scatter_kws = {'alpha':0.6}, label = 'developed countries')
sns.regplot(data =  df_developing, x = 'sdi', y = 'urban_population', fit_reg = True, scatter_kws = {'alpha':0.6}, label = 'developing countries');

ax.set_title('Correlation between CO2 emissions and Urban Population' , pad = 30)
ax.set_xlabel('Urban population growth')
ax.set_ylabel('Sustainable development index')
ax.grid(False) 
ax.legend(loc = 'lower right' , bbox_to_anchor=(1.30, 0), frameon=False)


plt.show()
```

Here, we observe a strong negative correlation between sustainable development index and urban population in developing countries, and a weak negative correlation for that of developing countries. 

### Research question 5:  Which country has the highest per capita emissions of co2?

```python
#create a column to evaluate gni per capita by co2 emissions
df_data_18['gni/co2'] = (df_data_18['gni_per_capita'] / df_data_18['co2_emission'])
df_data_18.head()
```

```python
df_data_co2_top = df_data_18.groupby('country').max()['gni/co2'].sort_values( ascending = False)
top_10co2 = df_data_co2_top.head(10)
```

African countries tops the chart in carbon emissions relative to gni per capita distribution. This implies the seriousness of climate change in Africa as the continent do not have the capacity to curtail co2 emissions due to slow development.

### Conclusions 
* We observed an increase in population from 1990 till 2018, and a concurrent increase in life expectancy and income europe central Asia and America.
